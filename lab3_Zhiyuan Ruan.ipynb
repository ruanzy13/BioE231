```python
%ls
%matplotlib inline
import sys,re,operator
import matplotlib.pyplot as plt
import numpy as np

infile = open("romeoandjuliet.txt")
text = infile.read().lower() # returns a version with all upper case characters replaced with lower case 
infile.close()
text = re.sub('[^a-z\ \']+', " ", text) #replaces anything that is not a lowercase letter, a space, or an apostrophe with a space
all_words = list(text.split()) #split sentence into separate words

word_of_frequency = {} #Creates an empty dictionary
name =None
frequency = 0
length=len(all_words)

i = 0 # index all_words with integer
while i < length:
    if all_words[i] not in word_of_frequency.keys():
        name = all_words[i]
        frequency+=1
        word_of_frequency[name]=frequency
        frequency = 0
    else:
        word_of_frequency[all_words[i]]+=1
    i+=1
#print(word_of_frequency)
sorted_dic = sorted(word_of_frequency.items(), key=operator.itemgetter(1),reverse = True) #sort the dictionary according to decreasing word count
#print(sorted_dic)
fig = plt.figure(figsize=(100,50),dpi=80)
axes=fig.add_subplot(1,1,1)
labels, ys = zip(*sorted_dic) # assign keys and values in directionay to labels, y-axis


ind = np.arange(len(labels))    # the x locations for the groups
width = 0.65      # the width of the bars: can also be len(x) sequence
p1 = plt.bar(ind, ys, width)
plt.ylabel('Frequency')
plt.xlabel('Words assigned by number of their ranks') # assign each word to a number of their rank
plt.xticks(np.arange(0, len(labels), 200))
plt.title('Frequency of words used in romeoandjuliet')
plt.yticks(np.arange(0, 700, 100))
plt.show()





        
        
        
        


    


    
```
